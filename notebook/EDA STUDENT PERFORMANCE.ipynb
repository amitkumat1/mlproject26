```python
for i in df.columns:
    if df[i].dtype != 'object':
        plt.boxplot(x=df[i])
        plt.title(i)
        plt.show()
```

```python
col = ['math_score', 'reading_score', 'writing_score', 'total_score']
for i in col:
    col = ['math_score', 'reading_score', 'writing_score', 'total_score']
    for i in col:
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)
        iqr = q3 - q1
        lw = q1 - 1.5 * iqr
        uw = q3 + 1.5 * iqr
        df = df.loc[(df[i] > lw) & (df[i] < uw)].copy()
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    lw = q1 - 1.5 * iqr
    uw = q3 + 1.5 * iqr
    df = df.loc[(df[i] > lw) & (df[i] < uw)].copy()
```
